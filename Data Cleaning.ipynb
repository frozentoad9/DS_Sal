```python
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])  #removing (Glassdoor Estimate)
salary
```

```python
remove_dk = salary.apply(lambda x: x.replace('K', '').replace('$', '').strip())   #removing dollar, K symboladn White spaces
remove_dk
```

```python
remove_dk.value_counts()                    #converting $17-$23 per hour to $149K-$201K
```

```python
remove_dk = remove_dk.replace('17- 23 Per Hour', '149-201')           #replacing the values
remove_dk
```

```python
remove_dk.value_counts()
```

```python
df['min_salary'] = remove_dk.apply(lambda x: int(x.split('-')[0]))           #Creating new column for max and min salary
df['max_salary'] = remove_dk.apply(lambda x: int(x.split('-')[1])) 
df['Avg_Salary'] = (df.min_salary + df.max_salary)/2                         #Label for the given data (i.e. predicted salary)
```
